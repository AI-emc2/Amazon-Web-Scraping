```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = "https://www.amazon.in/s?k=playstation+5&crid=3E64LC73NBSNO&sprefix=playstation+5%2Caps%2C647&ref=nb_sb_noss_1"
```

```python
# Headers for request
HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36', 'Accept-Language': 'en-US, en;q=0.5'}) #add your user agent
```

```python
requests.get(URL, headers=HEADERS)
```

```python
webpage = requests.get(URL, headers=HEADERS)
print(webpage)
soup = BeautifulSoup(webpage.content, "html.parser")
links = soup.find_all("a", attrs={'class': 'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})
```

```python
product_data = []

for link in links:
    try:
        product_link = "https://amazon.in" + link.get('href')
        new_webpage = requests.get(product_link, headers=HEADERS)
        # print(new_webpage)
    except requests.exceptions.RequestException as e:
        print(f"Error accessing product link: {product_link}")
        # print(e)
    new_soup = BeautifulSoup(new_webpage.content, "html.parser")

    title = new_soup.find("span", attrs={"id": "productTitle"}).text.strip()

    try:
        price = new_soup.find("span", attrs={"class": 'a-price-whole'}).text.strip()
    except AttributeError:
        price = "Price not available"

    try:
        rating = new_soup.find("span", attrs={"class": 'a-icon-alt'}).text.strip()
    except AttributeError:
        rating = "Rating not available"

    product_data.append({
        'Product_name': title,
        'Price': price,
        'Rating': rating
    })

# Convert data to DataFrame
Product_price = pd.DataFrame(product_data)
```

```python
def strip_(str):
  return str.split()[0]
```

```python
# Assuming `Product_price` is your DataFrame
Product_price['Primary'] = Product_price.groupby('Product_name').ngroup() + 1
Product_price['Rating'] = Product_price["Rating"].apply(lambda x: strip_(x))
Product_price
```

```python
# Function to extract customer review data from a product page
def function_1(url, product_name):
    response = requests.get(url, headers=HEADERS)
    soup = BeautifulSoup(response.content, 'html.parser')

    customer_data = []

    review_divs = soup.find_all("div", attrs={"class": "a-section review aok-relative"})

    for review_div in review_divs:
        profile_name = review_div.find("span", class_="a-profile-name").get_text(strip=True)


        try:
            verified_purchase = review_div.find("span", class_="a-size-mini a-color-state a-text-bold").get_text(strip=True)
        except AttributeError:
            verified_purchase = "Verification status not available"

        customer_data.append({
            "Product Name": product_name,
            "Customer Name": profile_name,
            "Verified Purchase": verified_purchase
        })

    return customer_data
```

```python
list_1 = []

# Iterate through each product link
for link in links:
    try:
        product_link = "https://amazon.in" + link.get('href')
        new_webpage = requests.get(product_link, headers=HEADERS)
        new_soup = BeautifulSoup(new_webpage.content, "html.parser")
        product_name = new_soup.find("span", attrs={"id": "productTitle"}).text.strip()

        product_reviews = function_1(product_link, product_name)
        list_1.extend(product_reviews)
    except Exception as e:
        print(f"Error processing {product_link}: {e}")
        continue

df = pd.DataFrame(list_1)
```

```python
# Function to extract customer review data from a product page
def extract_review_data(url, product_name):
    response = requests.get(url, headers=HEADERS)
    soup = BeautifulSoup(response.content, 'html.parser')

    customer_data = []

    review_divs = soup.find_all("div", attrs={"class": "a-section review aok-relative"})

    for review_div in review_divs:
        profile_name = review_div.find("span", class_="a-profile-name").get_text(strip=True)

        try:
            review_rating = review_div.find("i", class_="a-icon-star").find("span", class_="a-icon-alt").get_text(strip=True)
        except AttributeError:
            review_rating = "Rating not available"

        review_date = review_div.find("span", class_="a-size-base a-color-secondary review-date").get_text(strip=True)

        try:
            review_text = review_div.find("span", class_="review-text").get_text(strip=True)
        except AttributeError:
            review_text = "Review text not available"

        try:
            verified_purchase = review_div.find("span", class_="a-size-mini a-color-state a-text-bold").get_text(strip=True)
        except AttributeError:
            verified_purchase = "Verification status not available"

        customer_data.append({
            "Product Name": product_name,
            "Rating": review_rating,
            "Review Date": review_date,
            "Review Text": review_text,
        })

    return customer_data
```

```python
all_product_reviews = []

# Iterate through each product link
for link in links:
    try:
        product_link = "https://amazon.in" + link.get('href')
        new_webpage = requests.get(product_link, headers=HEADERS)
        new_soup = BeautifulSoup(new_webpage.content, "html.parser")
        product_name = new_soup.find("span", attrs={"id": "productTitle"}).text.strip()

        product_reviews = extract_review_data(product_link, product_name)
        all_product_reviews.extend(product_reviews)
    except Exception as e:
        print(f"Error processing {product_link}: {e}")
        continue

all_product_reviews_df = pd.DataFrame(all_product_reviews)
```

```python
all_product_reviews_df['Primary'] = all_product_reviews_df.groupby('Product Name').ngroup() + 1
all_product_reviews_df['Rating'] = all_product_reviews_df["Rating"].apply(lambda x: strip_(x))
all_product_reviews_df
```

```python
review_dates = all_product_reviews_df['Review Date']
date_pattern = r'on (\d+\s\w+\s\d{4})'
dates = review_dates.str.extract(date_pattern)
all_product_reviews_df['Review Date'] = pd.to_datetime(dates[0], format='%d %B %Y')
all_product_reviews_df.info()
```
